```python
import pandas as pd
from utils import connect_ssh_tunnel, connect_to_db, relative_path

config_file = relative_path("config_vm.yaml")

# Se connecter à la base de données MongoDB
sshtunnel = connect_ssh_tunnel(config_file, "ssh_mysql")
mysqlEngine = connect_to_db(config_file, "database_mysql")


mysqlConn = mysqlEngine.connect().execution_options(isolation_level="AUTOCOMMIT")
mysqlConn.begin()

df_messages = pd.read_sql('Messages', mysqlConn)
df_users    = pd.read_sql('Users'   , mysqlConn)
df_notes    = pd.read_sql('Notes'   , mysqlConn)
df_course   = pd.read_sql('Course'  , mysqlConn)
df_threads  = pd.read_sql('Threads' , mysqlConn)

sshtunnel.close()
```
